```python
import pandas as pd
import seaborn as sms


vendas = pd.read_excel("varejo.xlsx")

vendas.head()

```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_")
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.info()
```

```python
#filtrat categorias nulas

vendas[vendas["estado"].isnull()]
```

```python
#função para retirar as linhas com valores nulos

vendas_sem_nulos= vendas.dropna()
vendas_sem_nulos.info()
```

```python
#trocar linhas vazias por um valor

vendas["estado"].fillna("MS", inplace=True)
vendas.info()
```

```python
# trocar as linhas vazias por um valor médio

media_preço= vendas["Preço"].mean()
vendas["Preço"].fillna(media_preço,inplace = True)

vendas.info()
```

```python
vendas[["Preço","Preço_com_frete"]].describe()
```

```python
#corrigindo a base para tirar os valores que não fazem sentido
preço_errado = vendas.query("Preço > Preço_com_frete")
vendas_corretas = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_corretas.describe()
```

```python
vendas_corretas.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
#departamentos mais vendidos

vendas_corretas.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(vendas_corretas.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False),2)
```

```python
#quantidade de vendas por mês


vendas_corretas.groupby("Data").idcompra.nunique().sort_values(ascending=False)
```

```python
#meses com maiores vendas

vendas_corretas['mes'] = vendas_corretas["Data"].dt.month
vendas_corretas.groupby("mes").idcompra.nunique().sort_values(ascending=False)
```

```python
# fazer o join entre duas tabelas

cliente = pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
# fazendo Join das tabelas

cliente[["idade", "renda"]].describe()
cliente.info()
```

```python
#transformar "renda" em float

cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
#fazer o join

vendas_cliente = vendas_corretas.merge(cliente, how= "left", on="cliente_Log")
vendas_cliente.head()
```

```python
# média de renda por canal de renda
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").reset_index(),2)

agg_idcanal_renda
```

```python
#média de idade de clientes por bandeira

agg_idade_bandeira= round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2 )
agg_idade_bandeira
```

```python
#Vizualização usando matplotlib.pyplot

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5,4))

plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="green")
plt.ylabel("Média de idade")
plt.title("Idade média por Bandeira", loc="left",fontsize=18, fontweight="bold", color = "black")
```
